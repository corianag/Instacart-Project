```python
def price_label(row):
    
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Applying the function

df['price_range'] = df.apply(price_label, axis = 1)
```

#Creating a new column, “busiest day,” that will contain one of three different values: “Busiest day,” “Least busy,” and “Regularly busy.” 

```python
result = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")

```

```python
result
```

```python
#Creating a new column within ords_prods_merge dataframe and set it equal to result

df_ords_prods_merged['busiest_day'] = result
```

```python
#Printing the frequency of the new column created above


df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
#Checking the new column created above

df_ords_prods_merged.head()
```

#Creating a new column: "Busiest days". The busiest days are Saturday and Sunday and the lowest are Tuesday and Wednesday

```python
#Creating empty shell and adding labels:

result1 = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        result1.append("Busiest days")
    elif value == 3 or value == 4:
        result1.append("Slowest days")
    else:
        result1.append("Regularly days")
```

```python
result1
```

```python
#Creating the column 'busiest days'

df_ords_prods_merged['busiest days'] = result1
```
